Download a sample NetCDF dataset.

```python
url = 'https://drive.google.com/file/d/1EqgSCXNLGAvQgOdXJnt5T2QGEfBMxhYy/view?usp=sharing/wrf_dTW_2023-10-22_13_00_00.nc'
filename = 'wrf_dTW_2023-10-22_13_00_00.nc'
```

```python
leafmap.download_file(url, output=filename,  overwrite=True)
```

Read the NetCDF dataset.

```python
data = leafmap.read_netcdf(filename)
data
```

```python
import xarray as xr
import leafmap

# Load the dataset
ds = xr.open_dataset(filename)

# Select a specific level
selected_level = 1000.0  # Choose the desired level
ds_selected = ds.sel(level=selected_level)

# Extract the correct latitude and longitude values
lat_values = ds_selected['LAT'].values[:, 0]
lon_values = ds_selected['LON'].values[0, :]

# Assign these values to the lat and lon dimensions
ds_selected = ds_selected.assign_coords({"lat": lat_values, "lon": lon_values})

# Drop the original LAT and LON variables to avoid confusion
ds_selected = ds_selected.drop_vars(["LAT", "LON"])

# Rename the dimensions for compatibility with leafmap
ds_selected = ds_selected.rename({"south_north": "lat", "west_east": "lon"})

# Save the modified dataset to a temporary NetCDF file
temp_filename = "temp_wind_data.nc"
ds_selected.to_netcdf(temp_filename)

# Create a leafmap Map object and visualize the wind data
m = leafmap.Map(layers_control=True)
m.add_basemap('CartoDB.DarkMatter')
m.add_velocity(temp_filename, zonal_speed='U', meridional_speed='V')
m
```

```python
import xarray as xr
import leafmap

# Load the dataset
ds = xr.open_dataset(filename)

# Select a specific level (if necessary)
selected_level = 1000.0  # Choose the desired level
ds_selected = ds.sel(level=selected_level)

# Extract temperature data and coordinates
temperature = ds_selected['T'].values
lat_values = ds_selected['LAT'].values[:, 0]
lon_values = ds_selected['LON'].values[0, :]

# Prepare data for heatmap
heatmap_data = []
for i in range(len(lat_values)):
    for j in range(len(lon_values)):
        heatmap_data.append([lat_values[i], lon_values[j], temperature[i, j]])

# Create a leafmap Map object
m = leafmap.Map(layers_control=True)
m.add_basemap('CartoDB.DarkMatter')

# Add the heatmap
m.add_heatmap(heatmap_data, name="Temperature", radius=20)

# Display the map
m
```
